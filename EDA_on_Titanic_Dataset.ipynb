```python
# Task 2: Exploratory Data Analysis (EDA)

# ==============================
# Step 1: Importing Libraries
# ==============================
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px

# For inline plotting
%matplotlib inline

# ==============================
# Step 2: Loading Dataset
# ==============================
# Titanic dataset (can be uploaded manually or use seaborn's version)
df = sns.load_dataset("titanic")

# Show first few rows
print("First 5 rows of dataset:")
print(df.head())

# ==============================
# Step 3: Summary Statistics
# ==============================
print("\nDataset Info:")
print(df.info())

print("\nSummary Statistics:")
print(df.describe(include="all"))

# ==============================
# Step 4: Missing Values
# ==============================
print("\nMissing Values:")
print(df.isnull().sum())

# ==============================
# Step 5: Histograms for Numeric Features
# ==============================
df.hist(bins=20, figsize=(15, 10), color='skyblue', edgecolor='black')
plt.suptitle("Histograms of Numeric Features", fontsize=16)
plt.show()

# ==============================
# Step 6: Boxplots
# ==============================
plt.figure(figsize=(12,6))
sns.boxplot(x="pclass", y="age", data=df, palette="Set2")
plt.title("Boxplot of Age across Passenger Classes")
plt.show()

# ==============================
# Step 7: Correlation Matrix
# ==============================
plt.figure(figsize=(10,6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()

# ==============================
# Step 8: Pairplot
# ==============================
sns.pairplot(df[["age", "fare", "pclass", "survived"]].dropna(), hue="survived", palette="Set1")
plt.show()

# ==============================
# Step 9: Interactive Visualization (Plotly)
# ==============================
fig = px.scatter(df, x="age", y="fare", color="sex", size="fare", hover_data=["pclass", "survived"])
fig.update_layout(title="Age vs Fare (Interactive Plot)")
fig.show()

# ==============================
# Step 10: Insights
# ==============================
print("\n📌 Insights:")
print("- Most passengers are in 3rd class.")
print("- Females had a higher survival rate than males.")
print("- Younger passengers survived more often.")
print("- Higher fare passengers had higher survival probability.")
```
